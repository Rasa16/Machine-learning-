```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
import matplotlib as pl

%matplotlib inline
```

```python
train = pd.read_excel('F:\\study material\\covid 19 dataset\\Train_dataset.xlsx')
train.head(10)
```

```python
test = pd.read_excel('F:\\study material\\covid 19 dataset\\Test_dataset.xlsx')
test.head(10)
```

```python
train['Children'].value_counts().idxmax()
test['Children'].value_counts().idxmax()

train['Children'].fillna('2.0',inplace = True)
test['Children'].fillna('2.0',inplace = True)

train['Occupation'].value_counts().idxmax()
test['Occupation'].value_counts().idxmax()

train['Occupation'].fillna('Legal',inplace = True)
test['Occupation'].fillna('Legal',inplace = True)


train['Mode_transport'].value_counts().idxmax()
test['Mode_transport'].value_counts().idxmax()

train['Mode_transport'].fillna('Public',inplace = True)
test['Mode_transport'].fillna('Public',inplace = True)

train['cardiological pressure'].fillna('Normal',inplace = True)
test['cardiological pressure'].fillna('Normal',inplace = True)

train['Diuresis'].fillna('276.0',inplace = True)
test['Diuresis'].fillna('276.0',inplace = True)

train['Platelets'].fillna('88.0',inplace = True)
test['Platelets'].fillna('88.0',inplace = True)

train['HBB'].fillna('117.0',inplace = True)
test['HBB'].fillna('117.0',inplace = True)

train['d-dimer'].fillna('276.0',inplace = True)
test['d-dimer'].fillna('276.0',inplace = True)

train['Heart rate'].fillna('75.0',inplace = True)
test['Heart rate'].fillna('75.0',inplace = True)

train['HDL cholesterol'].fillna('53.0',inplace = True)
test['HDL cholesterol'].fillna('53.0',inplace = True)

train['Insurance'].fillna('2600000.0',inplace = True)
test['Insurance'].fillna('2600000.0',inplace = True)

train['FT/month'].fillna('1.0',inplace = True)
test['FT/month'].fillna('1.0',inplace = True)
```

```python
dummies = pd.get_dummies(train['comorbidity'])
dummies1 = pd.get_dummies(test['comorbidity'])

train = pd.concat([train,dummies],axis='columns')
test = pd.concat([test,dummies1],axis='columns')

train.drop('comorbidity',axis='columns',inplace=True)
test.drop('comorbidity',axis='columns',inplace=True)

train.drop('None',axis='columns',inplace=True)
test.drop('None',axis='columns',inplace=True)
```

```python
dummies = pd.get_dummies(train['Mode_transport'])
dummies1 = pd.get_dummies(test['Mode_transport'])

train = pd.concat([train,dummies],axis='columns')
test = pd.concat([test,dummies1],axis='columns')

train.drop('Mode_transport',axis='columns',inplace=True)
test.drop('Mode_transport',axis='columns',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le_married = LabelEncoder()
le_occupation = LabelEncoder()
le_gender = LabelEncoder()

#train
train['Married_n'] = le_married.fit_transform(train['Married'])
train['Occupation_n'] = le_occupation.fit_transform(train['Occupation'])
train['Gender_n'] = le_gender.fit_transform(train['Gender'])

#test
test['Married_n'] = le_married.fit_transform(test['Married'])
test['Occupation_n'] = le_occupation.fit_transform(test['Occupation'])
test['Gender_n'] = le_gender.fit_transform(test['Gender'])
```

```python
train.drop('Married',axis='columns',inplace=True)
test.drop('Married',axis='columns',inplace=True)

train.drop('Occupation',axis='columns',inplace=True)
test.drop('Occupation',axis='columns',inplace=True)

train.drop('Gender',axis='columns',inplace=True)
test.drop('Gender',axis='columns',inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
scaler = MinMaxScaler()

#train
scaler.fit(train[['Insurance']])
train['Insurance'] = scaler.transform(train[['Insurance']])

scaler.fit(train[['salary']])
train['salary'] = scaler.transform(train[['salary']])

#test
scaler.fit(test[['Insurance']])
test['Insurance'] = scaler.transform(test[['Insurance']])

scaler.fit(test[['salary']])
test['salary'] = scaler.transform(test[['salary']])
```

```python
train.replace({'<100':0, '<200':1, '<300':2, '<400':3},inplace = True)
train.replace({'Elevated':0, 'Normal':1, 'Stage-01':2, 'Stage-02':3},inplace = True)


test.replace({'<100':0, '<200':1, '<300':2, '<400':3},inplace = True)
test.replace({'Elevated':0, 'Normal':1, 'Stage-01':2, 'Stage-02':3},inplace = True)
```

```python
train = train.drop(['people_ID','Name','Designation'], axis = 1)
```

```python
train_x = train.drop(['Infect_Prob','Region'],axis = 1)

train_y = train.Infect_Prob
```
